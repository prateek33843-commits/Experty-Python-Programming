# Encapsulation Only:

**Question :** Design a class BankAccount where balance cannot be accessed directly.

```python
# create a class BankAccount.
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance #Private Attributes Balance

  # Deposite function.
  def deposite(self, amount):
    if amount > 0:
      self.__balance += amount
      return "Successfully deposited"
    elif amount < 0:
      return "Invalid amount"
    else:
      self.__balance
      return "Successfully deposited"

  # Withdraw function.
  def withdraw(self, amount):
    if amount > 0:
      self.__balance -= amount
      return "Successfully withdraw"
    elif amount < 0:
      return "Invalid amount"
    else:
      self.__balance
      return "Successfully withdraw"

  # Check Our balance.
  def get_balance(self):
    return self.__balance

acc = BankAccount(100)
```

```python
acc.deposite(200)
```

```python
acc.withdraw(150)
```

```python
acc.get_balance()
```

# Inheritance Only:
**Question :** Reuse behavior from a base class.

```python
# Create a class  for Vehicle.
class Vehicle():
  def __init__(self,model,brand):
    self.model = model
    self.brand = brand

# subclass inherite from the parent class.
class type(Vehicle):
  def __init__(self,model,brand,mileage,vehicle_type):
    super().__init__(model,brand)
    self.mileage = mileage
    self.vehicle_type = vehicle_type
```

```python
vehicle1 = type("M5","BMW","21km/L","Car")
print("Vehicle_Model:",vehicle1.model)
print("Vehicle_Brand:",vehicle1.brand)
print("Vehicle_Mileage:",vehicle1.mileage)
print("Vehicle_Type:", vehicle1.vehicle_type)
```

# Encapsulation & Polymorphism :
**Question :** Different authentication mechanisms with hidden credentail.

```python
# Create a class for Login by password
class Login:
  def __init__(self, password):
    self.__password = password # password is a private attribute

  def _authenticate(self, input_val):
    if self.__password == input_val:
      return "Correct password ! Login successfully."

    else:
      return "Incorrect password ! Try again."

# Create a child class for OTPLogin
class OTPLogin(Login):
  def __init__(self, set_otp):
    self.__otp = set_otp # Set_OTP is a private attribute

  def authenticate(self, otp):
    if self.__otp == otp:
      return "OTP Verified ! Login successfully."

    else:
      return "Incorrect OTP ! Try again."

# creating instances of class and subclass.
user_login = Login("Pass123")
OTP_login = OTPLogin("2589")
```

```python
user_login._authenticate("Pass123")
```

```python
OTP_login.authenticate("2589")
```

# Abstraction & Inheritance :
**Question :** Abstract shape reused by concrete shapes.

```python
from abc import ABC, abstractmethod

# Create a abstract class for shape
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

# create a subclass for circle
class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    print(f"Area of circle is : {3.14 * self.radius * self.radius}")
```

```python
# set a radius of cricle
my_circle = Circle(6)
# Find a area of my_circle
my_circle.area()
```

# Encapsulation, Inheritance & Abstraction :
**Question :** Account system with hidden balance and abstract interest logic.

```python
# import ABC
from abc import ABC, abstractmethod

# Create a parent class for Account
class Account(ABC):
  def __init__(self,balance):
    self.__balance = balance # balance is a private attribute

# get method for show our private attribute
  def _get_balance(self):
    return self.__balance

# set method for update our private attribute
  def _set_balance(self,balance):
    self.__balance = balance

  @abstractmethod
  def calculate_interest(self):
    pass

# Create a child class for Saving Account
class SavingAccount(Account):
  def calculate_interest(self): # create a function to calculate interest of Saving accounts
    new_balance = self._get_balance() * 1.04 # Saving account gets 4% interest.
    self._set_balance(new_balance)
    print(f'Saving account interest applied. New balance is {self._get_balance():.2f}')
```

```python
my_saving = SavingAccount(10000)
print(f"My Saving account balance is {my_saving._get_balance()}")
```

```python
my_saving.calculate_interest()
```

# Abstraction, Inheritance & Polymorphism :
**Question :** Notification System.

```python
# Import ABC
from abc import ABC, abstractmethod

# Create a class for notification
class Notification(ABC):
  def __init__(self,recipient,message):
    self.recipient = recipient
    self.message = message

  @abstractmethod
  def send(self):
    pass

# Create a sub class for email notification
class Email(Notification):
  def send(self):
    print(f'Email sent to {self.recipient}: {self.message}')

# Create a sub class for SMS notification
class SMS(Notification):
  def send(self):
    print(f'SMS sent to {self.recipient}: {self.message}')

# creating instances of subclass.
email_alert = Email("adesh@example.com", "Your EPP assignment is due soon.")
SMS_alert = SMS("+91-9876543210", "Your System is Online.")
```

```python
email_alert.send()
```

```python
SMS_alert.send()
```

# All 4 Pillars (Complete OOP) :
**Question :** Payment Processing System.


```python
# Import ABC
from abc import ABC, abstractmethod

# Create a class for payment
class Payment(ABC):
  def __init__(self, amount, recipient):
    self.__amount = amount
    self.__recipient = recipient # Encapsulation

# create get method for show private attributes

  def get_recipient(self):
    return self.__recipient

  def get_amount(self):
    return self.__amount

  @abstractmethod
  def process_payment(self): # Abstraction
    pass

# Create a sub class for Credit Card Payment

class CreditCardPayment(Payment):# Inheritance
  def process_payment(self):
    print(f'Paid RS. {self.get_amount()} to {self.get_recipient()} via Credit Card.')

# Create a sub class for UPI Payment

class UPIPayment(Payment):
  def process_payment(self): # Polymorphism
    print(f'Paid RS. {self.get_amount()} to {self.get_recipient()} via UPI')

# creating instances of subclass.
credit_payment = CreditCardPayment(10000, "Flipkart")
upi_payment = UPIPayment(50000, "Adesh Kumar Singh")
```

```python
credit_payment.process_payment()
```

```python
upi_payment.process_payment()
```
