Question 1. Create an ElectricCar class that interits from the car class and includes and additional attribute,battery_size.

```python
class Car:
  def __init__(self,model,brand):
    self.model=model
    self.brand=brand

class type(Car):
  def __init__(self,model,brand,battery_size):
    super().__init__(model,brand)

    self.battery_size=battery_size

car=type("M3","BMW","70KW")
print("Car Model:",car.model)
print("Car Brand:",car.brand)
print("Car Battery Size:",car.battery_size)
```

Question 2: Define a Laptop class that interits from the Computer class and includes an additional attribute,battery_life.

```python
class Computer:
  def __init__(self,model,brand):
    self.model=model
    self.brand=brand

class Laptop(Computer):
  def __init__(self,model,brand,battery_life):
    super().__init__(model,brand)
    self.battery_life=battery_life

computer=Laptop("82XM","Lenovo","11 Hours")
print("Computer Model:",computer.model)
print("Computer Brande:",computer.brand)
print("Computer battery Life:",computer.battery_life)
```

Question 3: Create a Smartphone class and add anothor chlid class of your choice that inherits from an ElectronicDevice class. Include an additional attribute, operating_system.

```python
class Smartphone:
  def __init__(self,model,brand,storage):
    self.model=model
    self.brand=brand
    self.storage=storage

class ElectronicDevice(Smartphone):
  def __init__(self,model,brand,storage,operating_system):
    super().__init__(model,brand,storage)
    self.operating_system=operating_system

sm=ElectronicDevice("16 pro","I Phone","2 Tb","OIS")
print("Smart Phone Model:",sm.model)
print("Smart Phone Brand:",sm.brand)
print("Smart Phone Storage:",sm.storage)
print("smart Phone Operating System:",sm.operating_system)
```

Question 4: Implement a Musician class and create another child class of your choice that extends an artist class. Include an additional attribute,instrument_played.  

```python
class Musician:
  def __init__(self,name,city):
    self.name=name
    self.city=city

class Artist(Musician):
  def __init__(self,name,city,instrument_played):
    super().__init__(name,city)
    self.instrument_played=instrument_played

info=Artist("madhurima","Amethi","Guitar")
print("Musician Name:",info.name)
print("Musician city:",info.city)
print("Musician Instrument:",info.instrument_played)
```

Question 5: Define a FictionBook class and add another child class of your choice that inherits from the Book class. Include an additional attribute,genre.

```python
class FictionBook:
  def __init__(self,book,writer):
    self.book=book
    self.writer=writer

class F_Book(FictionBook):
  def __init__(self,book,writer,genre):
    super().__init__(book,writer)
    self.genre=genre

Book=F_Book("Atomic Habits","James Clear","Self_help")
print("Book  Name:",Book.book)
print("Book Writer:",Book.writer)
print("Book genre:",Book.genre)
```

Bonus Question:- Create a student class that takes the name & marks of 3 subject as arguments in the constructor. Then create a method to print the average.

```python
class Students:
  def __init__(self,name,marks1,marks2,marks3):
    self.name=name
    self.marks1=marks1
    self.marks2=marks2
    self.marks3=marks3

  def average(self):
    average_marks=(self.marks1 + self.marks2 + self.marks3) / 3
    return average_marks

student=Students("Madhurima Singh",82,90,79)
print(f"Student Name:{student.name} And Average Marks Is: {student.average()}")
```
