```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

Queation 1 :- Basic web scaping without header

```python
url="https://www.scrapethissite.com/pages/simple/"
response=requests.get(url)

if response.status_code==200:
  print("success")

else:
  print("Failed")
```

```python
soup=BeautifulSoup(response.text,"html.parser")
print(soup.prettify())
```

```python
headline = soup.find("h1").text.strip()
description = soup.find("p", class_ = "lead").text.strip()
print("\nHeadline :-")
print(headline)
print("\nDescription :-")
print(description)
scrapped_data = []
country_block = soup.find_all("div", class_ = "col-md-4 country")
for block in country_block:
  name = block.find("h3", class_ = "country-name").text.strip()
  capital = block.find("span", class_ = "country-capital").text.strip()
  population = block.find("span", class_ = "country-population").text.strip()
  area = block.find("span", class_ = "country-area").text.strip()
  scrapped_data.append({"Name-": name,
                        "Capital-": capital,
                        "Population-": population,
                        "Area-": area})
print("\nScrapped_data :-")
for data in scrapped_data:
  print(data)
```

```python
url="https://www.goodreads.com/"

header={"user-Agent" : "Mozilla/5.0"}

response=requests.get(url,headers=header)

if response.status_code==200:
  print("Success")

else:
  print("Failed")
```

```python
soup=BeautifulSoup(response.text,"html.parser")

print(soup.prettify())
```

```python
quotes=soup.find_all("div",class_ = "quoteText")
authors=soup.find_all("span",class_ = "authorOrTitle")

for quote,author in zip(quotes,authors):
  print(f"{author.text.strip()} - {quote.text.strip()}")
```

```python
def Book_titles():
  url="http://books.toscrape.com"
  response=requests.get(url)
  soup=BeautifulSoup(response.text,"html")

  books=soup.find_all("h3")[:10]
  for i, book in enumerate(books,1):
    title=book.a["title"]
    print(f"{i}.{title}")

Book_titles()
```

```python
def get_bbc_headlines():
  url="https://www.bbc.com/news"
  header={"user-Agent":"Mozilla/5.0"}
  response=requests.get(url,headers=header)

  soup=BeautifulSoup(response.text,"html.parser")
  headlines=soup.find_all("h2")

  for i,h in enumerate(headlines[:5]):
    print(f"{i+1}.{h.get_text(strip = True)}")

get_bbc_headlines()
```

```python
url="https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/146.0.0.0 Safari/537.36"}
response=requests.get(url,headers=headers)
response.raise_for_status()
soup=BeautifulSoup(response.text,"html.parser")
tables=soup.find_all("table",class_ ="wikitable")
if not tables:
  raise ValueError("No table with class'wikitable' found on the page.")
table=tables[0]
rows=table.find_all("tr")
data=[]
for row in rows:
  cells=row.find_all(["td","th"])
  cells=[cell.get_text(strip=True) for cell in cells]
  data.append(cells)
df=pd.DataFrame(data)
print(df.head())
```
