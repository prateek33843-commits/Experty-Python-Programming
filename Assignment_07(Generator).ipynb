# Generators

```python
def travel_gen():
  yield "Delhi"
  yield "Mumbai"
  yield 2026

my_trip = travel_gen()
```

```python
print(next(my_trip))
```

```python
def Web_startup():
  yield "System Booting"
  yield "Network Connected"
  yield "Database Connected"
  yield "Loading Preferences"
  yield "Online and Ready"

web_loading = Web_startup()

for step in web_loading:
  print(step)
```

```python

house_price = (area * 5000 for area in range(1000, 1600, 100))

print("1000 sq ft Price is:",next(house_price))
print("1100 sq ft Price is:",next(house_price))
print("1200 sq ft Price is:",next(house_price))
print("1300 sq ft Price is:",next(house_price))
print("1400 sq ft Price is:",next(house_price))
print("1500 sq ft Price is:",next(house_price))
```

```python
def factorial_series(n):
  result = 1
  for i in range(1, n+1):
    result *= i
    yield result

fact = factorial_series(10)

print(list(fact))
```

```python
def running_average():
    total = 0
    count = 0
    average = 0

    while True:
        new_value = (yield average)

        if new_value is not None:
            total += new_value
            count += 1
            average = total / count

avg_cal = running_average()

print("Initial state:", next(avg_cal))

print("After sending 10:", avg_cal.send(10))
print("After sending 20:", avg_cal.send(20))
print("After sending 30:", avg_cal.send(30))
print("After sending None:", avg_cal.send(None))
```
