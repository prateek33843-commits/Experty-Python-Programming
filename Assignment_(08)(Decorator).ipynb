# Logging Decorator (Fixed Arguments)

Question : Create a decorator that logs function name and parameters. The function should take exactly 2 arguments.

```python
def log_function(func):
  def wrapper(a,b):
    print(f"Callint {func.__name__} function , with argument{a},{b}")
    result = func (a,b)

    print(f"{func.__name__} Result:{result}")

  return wrapper
@log_function
def multiply(a,b):
  return a*b
c=multiply(9,8)
```

# 3. Authorization Decorator(Single string)

Question: Create a decorator that if a user is "admin".

```python
def required_admin(func):
  def wrapper(user):
    if user != "Admin":
      print("Access denied")
      return
    return func(user)

  return wrapper

@required_admin
def access_dashboard(user):
  print(f"Welcome {user} ,Your dashboard accessed")


access_dashboard("guest")
access_dashboard("Admin")
```

Create a memoization decorator for a function that takes one integer parameter.

# Returning Values

Decorators can also return values from the wrapped function:

```python
def double_result(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result*2
    return wrapper

@double_result
def get_number():
    return 60

print(get_number())
```
